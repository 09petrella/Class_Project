```python
import numpy as np
import pandas as pd
```

```python
import gzip
import shutil
with gzip.open('name.basics.tsv.gz', 'rb') as f_in:
    with open('./name.basic.txt', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
        
with gzip.open('title.akas.tsv.gz', 'rb') as f_in:
    with open('./title.akas.txt', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)

with gzip.open('title.basics.tsv.gz', 'rb') as f_in:
    with open('./title.basics.txt', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)

with gzip.open('title.ratings.tsv.gz', 'rb') as f_in:
    with open('./title.ratings.txt', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)

with gzip.open('title.principals.tsv.gz', 'rb') as f_in:
    with open('./title.principals.txt', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
```

```python
name_basics  = pd.read_csv('name.basic.txt',       sep='\t')
akas         = pd.read_csv('title.akas.txt',       sep='\t')
title_basics = pd.read_csv('title.basics.txt',     sep='\t')
principals   = pd.read_csv('title.principals.txt', sep='\t')
ratings      = pd.read_csv('title.ratings.txt',    sep='\t')
```
